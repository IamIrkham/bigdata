```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Memasukkan data ke dalam DataFrame
data = {
    'age': ['<=30', '<=30', '31...40', '>40', '>40', '>40', '31...40', '<=30', '<=30', '>40', '<=30', '31...40', '31...40', '>40'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

df = pd.DataFrame(data)

# Mengkodekan data kategorikal ke dalam bentuk numerik
df['age'] = df['age'].map({'<=30': 0, '31...40': 1, '>40': 2})
df['income'] = df['income'].map({'low': 0, 'medium': 1, 'high': 2})
df['student'] = df['student'].map({'no': 0, 'yes': 1})
df['credit_rating'] = df['credit_rating'].map({'fair': 0, 'excellent': 1})
df['buys_computer'] = df['buys_computer'].map({'no': 0, 'yes': 1})

# Membagi data menjadi fitur dan label
X = df[['age', 'income', 'student', 'credit_rating']]
y = df['buys_computer']

# Melatih model decision tree
model = DecisionTreeClassifier()
model.fit(X, y)

# Visualisasikan decision tree
plt.figure(figsize=(12,8))
tree.plot_tree(model, feature_names=['age', 'income', 'student', 'credit_rating'], class_names=['no', 'yes'], filled=True)
plt.show()
```
